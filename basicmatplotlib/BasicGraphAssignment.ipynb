```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_Op = pd.read_csv('C:/Users/micah/CTBA/M2/OpidsVA.csv')
df_Op.head()

fig,ax = plt.subplots()
ax.scatter(df_Op['FPOO-Rate'],df_Op['MedianHouseholdIncome'], alpha = 0.5, c = 'g') #I chose scatterplot to show if there was a strong correlation or not.
fig.suptitle('Opioid Overdoses vs. Median Household Income')
ax.xaxis.set_label_text('Opioid Overdoses')
ax.yaxis.set_label_text('Median Household Income')
plt.show() #There appears to be a correlation
```

```python
fig,ax = plt.subplots()
ax.scatter(df_Op['FPOO-Rate'],df_Op['FFHO-Rate'], alpha = 0.5, c = 'b') #I chose scatterplot to show the relationship between the two
fig.suptitle('Opioid Overdoses & Heroin Overdoses')
ax.xaxis.set_label_text('Opioid Overdoses')
ax.yaxis.set_label_text('Heroin Overdoses')
plt.show() #While there is some correlation, I wouldn't say it is strong
```

```python
df_Pr = pd.read_csv('C:/Users/micah/CTBA/M2/presidents.csv')
df_Pr.columns.values
df_Pr.set_index('State')
States = df_Pr['State']
States = dict(States.value_counts())

x = [i for i in States]
freq = [States[i] for i in States]

print (x)
print (freq)

fig,ax = plt.subplots()
ax.bar(x=x,height=freq, color = 'g', alpha = 0.75) #I created a frequency histogram using bar chart to show number of occurances per state
fig.suptitle('President Birth Places')
fig.set_size_inches(7,5)
ax.xaxis.set_label_text('State')
ax.yaxis.set_label_text('Presidents Born')
```
